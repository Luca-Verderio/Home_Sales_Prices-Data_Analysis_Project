# 03. Data wrangling, cleaning and consistency checks

## I. Create a 'df_clean' data set

```python
df_clean = df
```

## II. Missing values

```python
# Find missing values
df_clean.isnull().sum()
```

No missing values found.

## III. Duplicate values

```python
# Find duplicate values
df_dups = df_clean[df_clean.duplicated()]
```

```python
# Return df_dups
df_dups
```

No duplicate values found.

## IV. Mixed-type data

```python
# Check for mixed-type data
for col in df_clean.columns.tolist():
    weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_clean[weird]) > 0:
        print (col)
```

No mix-type data found.

## V. Data types

```python
# Display data types
df_clean.info()
```

Although some variables are qualitative, in this case they will be left as 'int' or 'float' so that they can be fed into a machine learning model. In fact, machine learning models require all input and output variables to be numeric.

```python
# Convert 'date' from 'object' to 'datetime'
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
# Change date format from yyyy-mm-dd to yyyy-mm
df_clean['date'] = df_clean['date'].dt.strftime('%Y-%m')
```

## VI. Drop unnecessary columns

```python
# Remove 'id' column
df_clean.drop('id', axis = 1, inplace = True)
```
